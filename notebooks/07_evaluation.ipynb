# 07 · Evaluation & paper figures
Assemble the comparison tables and final figures that go into the IEEE paper.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from dga.config import load_config, set_seed
cfg = load_config(); set_seed(cfg.seed)
```

```python
import numpy as np
from dga.config import PROJECT_ROOT
from dga import data as dga_data, conventional, clustering, evaluation, viz
Z = np.load(PROJECT_ROOT/'results'/'models'/'latent.npy')
index = np.load(PROJECT_ROOT/'results'/'models'/'feature_index.npy', allow_pickle=True)
df = dga_data.load_clean().loc[index]; duval = conventional.diagnose(df)['duval'].to_numpy()
```

## Master comparison table
Add rows for AE+KMeans, VAE+KMeans, raw+KMeans, IsolationForest, etc.

```python
res = clustering.fit(Z, cfg.clustering)
table = evaluation.comparison_table({
    'AE + ' + res.algorithm: {**evaluation.internal_metrics(Z, res.labels),
                              **evaluation.external_metrics(res.labels, duval)},
})
table.to_csv(PROJECT_ROOT/'results'/'tables'/'master_comparison.csv')
table
```

## Export final latent figure

```python
viz.set_paper_style()
fig = viz.plot_latent_2d(Z, res.labels, title='Discovered clusters')
viz.save_fig(fig, 'fig_latent_clusters'); fig
```
