# 02 · Preprocessing
Build the model-ready feature matrix and visualise the effect of the log1p + scaling pipeline.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from dga.config import load_config, set_seed
cfg = load_config(); set_seed(cfg.seed)
```

```python
from dga import data as dga_data
from dga.preprocessing import build_feature_matrix, select_features
df = dga_data.load_clean()
print('features:', select_features(cfg))
fm = build_feature_matrix(df, cfg)
fm.X.shape, fm.features
```

## Before vs after transform

```python
raw = df[fm.features].loc[fm.index]
fig, axes = plt.subplots(1,2, figsize=(9,3))
axes[0].hist(raw['H2'].clip(upper=2000), bins=50); axes[0].set_title('H2 raw (clipped view)')
axes[1].hist(fm.X[:, fm.features.index('H2')], bins=50); axes[1].set_title('H2 after log1p + scaling'); fig
```
