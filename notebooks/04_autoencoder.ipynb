# 04 · Autoencoder representation learning
Learn a compact latent code for each sample; inspect training and the latent geometry.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from dga.config import load_config, set_seed
cfg = load_config(); set_seed(cfg.seed)
```

```python
from dga import data as dga_data
from dga.preprocessing import build_feature_matrix
from dga.models.autoencoder import train_autoencoder
df = dga_data.load_clean(); fm = build_feature_matrix(df, cfg)
res = train_autoencoder(fm.X, cfg.autoencoder)
Z = res.model.encode(fm.X)
Z.shape
```

## Training curves

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(res.history['train'], label='train')
ax.plot(res.history['val'], label='val'); ax.set_xlabel('epoch'); ax.set_ylabel('MSE'); ax.legend(); fig
```

## Latent space coloured by Duval fault
If the AE captures fault *type*, Duval classes should separate. If it captures *severity*, they will not — a key research question.

```python
from dga import conventional, viz
duval = conventional.diagnose(df).loc[fm.index, 'duval']
fig = viz.plot_latent_2d(Z, duval, title='AE latent (PCA) by Duval'); fig
```
